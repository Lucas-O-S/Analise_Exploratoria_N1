### 1.3. Corrigindo Dataset


#### 1.3.1. Links


O tipo de tmdbId está como float, como id não deve ter casa decimal vamos corrigir esse dado <br/>


```python
dfLinks.columns = ["Id de Filme","Id IMDB", "Id TMDB"]
```

```python
dfLinks["Id TMDB"] = dfLinks["Id TMDB"].fillna(0).astype(int)
```

Verificando resultados

```python
dfLinks.info()
```

```python
dfLinks.head()
```

#### 1.3.2. Movies


O tipo de Title e genres está como object, logo devemos corrigir estes valores: <br/>


```python
dfMovies.columns = ["Id de Filme","Titulo", "Genero"]
```

```python
dfMovies["Titulo"] = dfMovies["Titulo"].astype("string")
```

```python
dfMovies["Genero"] = dfMovies["Genero"].astype("string")
```

Verificando resultados

```python
dfMovies.info()
```

```python
dfMovies.head()
```

#### 1.3.3. Ratings


O tipo de timestamp está com valores ilegiveis logo deve ser corrigido: <br/>


```python
dfRatings.columns = ["Id de Usuario","Id de Filme", "Nota", "Data/Hora"]
```

```python
dfRatings["Data/Hora"] = pd.to_datetime(dfRatings["Data/Hora"], unit = "s")
```

Verificando resultados

```python
dfRatings.info()
```

```python
dfRatings.head()
```

#### 1.3.4. Tags


O tipo de tag está como object, logo devemos corrigir este valor, alem disso a timestamp está com valores ilegiveis: <br/>


```python
dfTags.columns = ["Id de Usuario","Id de Filme", "Tag", "Data/Hora"]
```

```python
dfTags["Tag"] = dfTags["Tag"].astype("string")
```

```python
dfTags["Data/Hora"] = pd.to_datetime(dfTags["Data/Hora"], unit = "s")
```
